```python
def plott(x, fig):
    plt.subplot(4,2,fig)
    plt.title(x+" Histogram")
    sns.boxplot(y = df[x])

    plt.subplot(4,2,(fig+1))
    plt.title(x+" Distplot")
    sns.distplot(df[x])

plt.tight_layout()
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(12,20))
plott('Pregnancies',1)
plott('Glucose',3)
plott('BloodPressure',5)
plott('SkinThickness',7)
plt.show()
```
